# 2. change data from text to vectors

```python
# import pretrained word2vec index
pretrained_word2vec_path = "DataSet\\GoogleNews-vectors-negative300.bin"
word2vec_pre = gs.models.KeyedVectors.load_word2vec_format(pretrained_word2vec_path, binary= True)
```

```python
def getVec(sentence):
    'function for getting vector of the sentence'
    word_list=sentence.split()
    sentence_matrix = list()
    for word in word_list:
        try:
            vec = word2vec_pre.word_vec(word)
            sentence_matrix.append(vec)
        except:
            pass
    sentence_matrix = np.array(sentence_matrix)
    return sentence_matrix
```

```python
# test
sen="i like musical res"
ma=getVec(sen)
print(ma)
```

```python
def fit_transform(input_dict):
    'function for transform the dictionary within data into dictionary within vector matrix'
    output_dict=dict()
    for each_key in input_dict.keys():
        sentences_matrix=list()
        for sentence in tqdm(input_dict[each_key]):
            sentence_vec = getVec(sentence)
            sentences_matrix.append(sentence_vec)
        sentences_matrix = np.array(sentences_matrix)
        output_dict[each_key]=sentences_matrix
    return output_dict
```

# 3.set different data format only based on no correction data for further test

```python

def normal(original_data):
    # get normal overall data
    length = len(original_data["turn1"])
    list_all = list()
    for i in tqdm(range(length)):
        list_turn1 = np.array(original_data["turn1"][i])
        list_turn2 =  np.array(original_data["turn2"][i])
        list_turn3 = np.array(original_data["turn3"][i])
        
        list_dialog = list(list_turn1)+list(list_turn2)+list(list_turn3)
        list_dialog = np.array(list_dialog)
        list_all.append(list_dialog)
    
    list_all = np.array(list_all)
    return list_all

def delete2nd(original_data):
    # delete 2nd turn dialog
    length = len(original_data["turn1"])
    list_all = list()
    for i in tqdm(range(length)):
        list_turn1 = np.array(original_data["turn1"][i])
        list_turn3 = np.array(original_data["turn3"][i])
        
        list_dialog = list(list_turn1)+list(list_turn3)
        list_dialog = np.array(list_dialog)
        list_all.append(list_dialog)
    list_all = np.array(list_all)
    return list_all

def double1st3rdturn(original_data):
    length = len(original_data["turn1"])
    list_all = list()
    for i in tqdm(range(length)):
        list_turn1 = np.array(original_data["turn1"][i])*2
        list_turn2 =  np.array(original_data["turn2"][i])
        list_turn3 = np.array(original_data["turn3"][i])*2
        
        list_dialog = list(list_turn1)+list(list_turn2)+list(list_turn3)
        list_dialog = np.array(list_dialog)
        list_all.append(list_dialog)
    
    list_all = np.array(list_all)
    return list_all

def time3_3rdturn(original_data):
    # make turn1 multiplied by 2, and turn3 multiplied by 3
    length = len(original_data["turn1"])
    list_all = list()
    for i in tqdm(range(length)):
        list_turn1 = np.array(original_data["turn1"][i])*2
        list_turn2 =  np.array(original_data["turn2"][i])
        list_turn3 = np.array(original_data["turn3"][i])*3
        
        list_dialog = list(list_turn1)+list(list_turn2)+list(list_turn3)
        list_dialog = np.array(list_dialog)
        list_all.append(list_dialog)
    list_all = np.array(list_all)
    return list_all

def time2_no2nd(original_data):
    # delete turn2 and make turn3 multiplied by 2
    length = len(original_data["turn1"])
    list_all = list()
    for i in range(length):
        list_turn1 = np.array(original_data["turn1"][i])
        list_turn3 = np.array(original_data["turn3"][i])*2
        
        list_dialog = list(list_turn1)+list(list_turn3)
        list_dialog = np.array(list_dialog)
        list_all.append(list_dialog)
    list_all = np.array(list_all)
    return list_all
```

```python
def get_different_dataformat(original_data):
    data = dict()
    data["overall"] = normal(original_data)
    # data["delete2nd"]=delete2nd(original_data)
    data["double13"]=double1st3rdturn(original_data)
    data["time3_3rd"]=time3_3rdturn(original_data)
    # data["time2_no2nd"]=time2_no2nd(original_data)
    return data
```

```python
def save_data(name,data):
    pickle_file = open('DataSet\\'+name,mode='wb')
    pickle.dump(data,pickle_file)
    pickle_file.close()
```

## 3.2.add emoji/emoticon

```python
# function for getting vector from pre-trained dict of vectors of emoji/emoticon by myself
def getVec_emo(one_emo,dict_vec_emo_aver=dataset["vec_emo"]):
    try:
        vec_emo = dict_vec_emo_aver[one_emo]
    except KeyError:
        vec_emo = None
    return vec_emo
```

```python
def transfer_one_emotext(input_list):
    # here we get the list which is constructed as [no_happened, emoji/emoticon]
    no_happened = input_list[0]
    emoji=input_list[1]
    
    #here is the process of processing the text
    vec_emo = getVec_emo(emoji)
    # then since there can be multi same emoji/emoticon in one sentence, I need to multiply the vectors to fit the number
    words_vec=list()
    for i in range(no_happened):
        words_vec.append(vec_emo)
    #print("length of words_vec:{a}".format(a=len(words_vec)))
    return words_vec
        

def combine_multi_emo(emo_list):
    # since there can be multi different emoji/emoticons in one sentence, I need to combine them together as sentence of vectors
    sentence_list=list()
    for emo_list_one in emo_list:
        words_vec = transfer_one_emotext(emo_list_one)
        sentence_list = sentence_list+words_vec
    #print("words of sentence:{a}".format(a=len(sentence_list)))
    return sentence_list

def emo_transform(input_dict):
    output_dict=dict()
    for key in input_dict.keys():
        list_out=list()
        for emo_list in tqdm(input_dict[key]):
            # we can have emo_list in input_dict[key] such as
            # [[no_happened, meaning_stored],[no_happened, meaning_stored]]
            # or nothing -> emo_list = []
            sentence_list=[]
            if emo_list !=[]: 
                sentence_list = combine_multi_emo(emo_list)
            else:
                pass
            #print("{a} emoji/emoticons in this sentence!".format(a=len(sentence_list)))
            list_out.append(sentence_list)
        
        output_dict[key]=list_out

    return output_dict
```

```python
# combine 3 turn data with their emoji/emoticon meanings together
def combine_3turn(text_dict,emo_dict):
    combined_dict=dict()
    for key in text_dict.keys():
        combined_list=list()
        length = len(text_dict[key])
        for i in range(length):
            new_sentence = list(text_dict[key][i])+list(emo_dict[key][i])
            # print(len(new_sentence))
            combined_list.append(new_sentence)
        
        combined_dict[key]=combined_list
        
    return combined_dict
```

```python
# combine 3 turn data into 1 turn data
def combine_method(text_content):
    length = len(text_content["turn1"])
    text_all_content = list()
    for i in tqdm(range(length)):
        text_row_all = text_content["turn1"][i] + text_content["turn2"][i] + text_content["turn3"][i]
        text_all_content.append(text_row_all)
    
    text_all_content = {"all":text_all_content}
    
    return text_all_content
```

```python
# transform dictionary of extracted emoji/emoticon with their meanings to vectors
train_emo_3  = emo_transform(dataset["train_emo_3"])
test_emo_3  = emo_transform(dataset["test_emo_3"])

# transform 3 turn data into vectors
vec_train_3_no_emo  = fit_transform(dataset["train_3_no_emo"])
vec_test_3_no_emo  = fit_transform(dataset["test_3_no_emo"])
```
