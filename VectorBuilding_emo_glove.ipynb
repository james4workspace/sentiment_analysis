# 2.Compute Vectors

## 2.1. Get Dict Version

```python
#function for turning dataframe into dict
def get_dict(df,keyvalue,content):
    list_key=list(df[keyvalue])
    list_content=list(df[content])
    dict_new = dict(zip(list_key,list_content))
    return dict_new
```

```python
dict_emo = get_dict(emo_df,"emoji/emoticon","key words")
```

## 2.2. Import Word Embedding Resource and build Transformation Function

```python
# import embedding resource glove pretrained data
GloVe_path="DataSet\\glove.840B.300d.txt"
```

```python
def load_glove(path):
    glove_dict = dict()
    words = list()
    with open(path, mode='r', encoding="utf-8") as vec_file:
        for line in tqdm(vec_file):
            values = line.split()
            word = values[0]
            vec = np.array(values[1:])
            glove_dict[word] = vec

    print("There are {0} distinct word vectors in this pretrained GloVe DataSet.".format(len(glove_dict)))
    return glove_dict
```

```python
glove_dict = load_glove(GloVe_path)
```

```python
def getVec(sentence):
    'function for getting vector of the sentence'
    word_list=sentence.split()
    sentence_matrix = list()
    for word in word_list:
        try:
            vec = glove_dict[word]
            vec = np.array(vec).astype(np.float)
            sentence_matrix.append(vec)
        except:
            pass
    sentence_matrix = np.array(sentence_matrix)
    return sentence_matrix
```

```python
# test
sen="i like musical res"
ma=getVec(sen)
print(ma)
```

```python
def fit_transform(input_dict):
    'function for transform the dictionary within data into dictionary within vector matrix'
    output_dict=dict()
    for each_key in tqdm(input_dict.keys()):
        vectors_perSentence=getVec(input_dict[each_key])
        output_dict[each_key]=vectors_perSentence
    return output_dict
```

```python
# test
sen={"first":"i like schoco","second":"what is your name"}
ma=fit_transform(sen)
print(ma)
```

```python
dict_vec_emo = fit_transform(dict_emo)
```

## 2.3. Compute the Vector of each Emoji/Emoticon based on Average Algorithm

```python
def cal_averageVec(input_dict):
    output_dict=dict()
    for each_emo in tqdm(input_dict.keys()):
        vec_sentence = np.array(input_dict[each_emo])
        length = len(vec_sentence)
        sum_sen = vec_sentence.sum(axis=0)
        aver_sen = sum_sen/length
        output_dict[each_emo]=aver_sen
    
    return output_dict
```

```python
#test
sen={"first":[[1,2,3],[4,5,6]],"second":[[7,8,9],[10,11,12],[13,14,15]]}
cal_sen = cal_averageVec(sen)
print(cal_sen)
```

```python
dict_vec_emo_aver = cal_averageVec(dict_vec_emo)
```

## 2.5. Vector of Emoji/Emoticon Storage
